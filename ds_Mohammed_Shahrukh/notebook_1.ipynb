```python
# =======================================================
# Mohammed Shahrukh  Web3 Trading DS Assignment
# Colab Link: <https://colab.research.google.com/drive/10zoCxjZx3JcGZnJ7Z5-6pBxd-5Awtt-u?usp=sharing>
# =======================================================

# 1) Imports & dirs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

Path("csv_files").mkdir(exist_ok=True)
Path("outputs").mkdir(exist_ok=True)

def save_plot(fig, name):
    fig.savefig(f"outputs/{name}", dpi=150, bbox_inches='tight')

# 2) Dataset URLs (use as-is or replace with your own Drive copies)
trader_url = "https://drive.google.com/file/d/1IAfLZwu6rJzyWKgBToqwSmmVYU6VbjVs/view?usp=sharing"
fg_url     = "https://drive.google.com/file/d/1PgQC0tO8XN-wqkNyghWc_-mnrYv_nhSf/view?usp=sharing"

def gdrive_to_download(url: str) -> str:
    file_id = url.split("/d/")[1].split("/")[0]
    return f"https://drive.google.com/uc?id={file_id}"

trader_fp = gdrive_to_download(trader_url)
fg_fp     = gdrive_to_download(fg_url)

# 3) Load raw
trades = pd.read_csv(trader_fp)
fg     = pd.read_csv(fg_fp)
print("Trader Data:", trades.shape)
print("Fear/Greed Data:", fg.shape)

# 4) Robust cleaning & timestamp detection
trades.columns = trades.columns.str.strip().str.lower().str.replace(" ", "_")
fg.columns     = fg.columns.str.strip().str.lower().str.replace(" ", "_")

print("Trades columns:", list(trades.columns))
print("FG columns:", list(fg.columns))

# --- detect trades timestamp column
candidates_exact = ['time','timestamp','datetime','executed_at','created_at','event_time','block_time','timestamp_ist']
candidates_contains = ['time','date']
time_col = next((c for c in candidates_exact if c in trades.columns), None)
if time_col is None:
    time_col = next((c for c in trades.columns if any(sub in c for sub in candidates_contains)), None)
if time_col is None:
    raise ValueError("Could not find a time-like column in trades. Columns were: " + ", ".join(trades.columns))

# parse trades time (auto seconds vs ms if numeric)
series = trades[time_col]
if pd.api.types.is_numeric_dtype(series):
    v = float(series.dropna().iloc[0])  # epoch ms > 1e11
    unit = 'ms' if v > 1e11 else 's'
    trades['time'] = pd.to_datetime(series, unit=unit, utc=True, errors='coerce')
else:
    trades['time'] = pd.to_datetime(series, utc=True, errors='coerce')

# --- FG date + sentiment
fg_date_col = 'date' if 'date' in fg.columns else next((c for c in fg.columns if 'date' in c), None)
if fg_date_col is None:
    raise ValueError("Could not find a date column in Fear & Greed dataset. Columns: " + ", ".join(fg.columns))

if pd.api.types.is_numeric_dtype(fg[fg_date_col]):
    v = float(fg[fg_date_col].dropna().iloc[0])
    unit = 'ms' if v > 1e11 else 's'
    fg['date'] = pd.to_datetime(fg[fg_date_col], unit=unit, utc=True, errors='coerce')
else:
    fg['date'] = pd.to_datetime(fg[fg_date_col], utc=True, errors='coerce')

if 'classification' in fg.columns:
    fg = fg.rename(columns={'classification':'sentiment'})
elif 'sentiment' not in fg.columns:
    s_col = next((c for c in fg.columns if 'class' in c or 'sentiment' in c), None)
    if s_col is None:
        raise ValueError("Could not find sentiment/classification column in Fear & Greed dataset.")
    fg = fg.rename(columns={s_col: 'sentiment'})

# cast common numerics if present
for c in ['execution_price','size','size_usd','size_tokens','leverage','closedpnl','closed_pnl','start_position']:
    if c in trades.columns:
        trades[c] = pd.to_numeric(trades[c], errors='coerce')

# keep valid time rows
trades = trades.dropna(subset=['time'])

# =======================================================
# === COLUMN MAPPING + FEATURES / AGGREGATION
# =======================================================
def pick(*names):
    """Return the first existing column name from candidates, else None."""
    for n in names:
        if n and n in trades.columns:
            return n
    return None

col_price    = pick('execution_price', 'price', 'fill_price')
col_sizeusd  = pick('size_usd', 'notional_usd', 'quote_value_usd')
col_sizetok  = pick('size_tokens', 'size_token', 'size', 'qty', 'quantity')
col_side     = pick('side', 'direction', 'buy_sell')
col_pnl      = pick('closed_pnl', 'closedpnl', 'pnl', 'realized_pnl')
col_leverage = pick('leverage', 'lev')

print("Mapped columns ->",
      "\n  price:", col_price,
      "\n  size_usd:", col_sizeusd,
      "\n  size_tokens:", col_sizetok,
      "\n  side:", col_side,
      "\n  pnl:", col_pnl,
      "\n  leverage:", col_leverage)

# Ensure numerics
for c in [col_price, col_sizeusd, col_sizetok, col_pnl, col_leverage]:
    if c:
        trades[c] = pd.to_numeric(trades[c], errors='coerce')

# --- FEATURES ---
# volume_usd: prefer provided USD notional; else tokens * price
if col_sizeusd:
    trades['volume_usd'] = trades[col_sizeusd].abs()
elif col_sizetok and col_price:
    trades['volume_usd'] = trades[col_sizetok].abs() * trades[col_price]
else:
    trades['volume_usd'] = np.nan

if col_pnl:
    trades['is_win'] = (trades[col_pnl] > 0).astype(int)
else:
    trades['is_win'] = np.nan

if col_side:
    s = trades[col_side].astype(str).str.lower()
    trades['is_long'] = s.isin(['buy','long']).astype(int)
else:
    trades['is_long'] = np.nan

# daily alignment
trades['trade_date'] = trades['time'].dt.floor('D')
fg['date'] = fg['date'].dt.floor('D')

# daily aggregate
agg_dict = {
    'n_trades':        ('account','count') if 'account' in trades.columns else ('time','count'),
    'accounts_active': ('account','nunique') if 'account' in trades.columns else None,
    'volume_usd':      ('volume_usd','sum'),
    'avg_leverage':    (col_leverage,'mean') if col_leverage else None,
    'win_rate':        ('is_win','mean'),
    'pnl_sum':         (col_pnl,'sum') if col_pnl else None,
    'avg_trade_size':  (col_sizetok,'mean') if col_sizetok else None,
    'long_share':      ('is_long','mean')
}
agg_dict = {k:v for k,v in agg_dict.items() if v is not None}
daily = (trades.groupby('trade_date').agg(**agg_dict).reset_index())

# merge with sentiment
df = daily.merge(fg[['date','sentiment']], left_on='trade_date', right_on='date', how='left').drop(columns=['date'])

# encode sentiment (support Neutral/Extreme Greed/Extreme Fear)
sent_map = {'NEUTRAL':0, 'FEAR':1, 'GREED':2, 'EXTREME GREED':3, 'EXTREME FEAR':-1}
df['sentiment_clean'] = df['sentiment'].astype(str).str.strip()
df['sentiment_code']  = df['sentiment_clean'].str.upper().map(sent_map)

# save core
df.to_csv("csv_files/daily_behavior_with_sentiment.csv", index=False)
print("Saved -> csv_files/daily_behavior_with_sentiment.csv")
display(df.head())

# sentiment comparison table
comp = (df.groupby('sentiment_clean')
        .agg(volume_usd=('volume_usd','mean'),
             avg_leverage=('avg_leverage','mean') if 'avg_leverage' in df.columns else ('volume_usd','mean'),
             win_rate=('win_rate','mean') if 'win_rate' in df.columns else ('volume_usd','mean'),
             pnl_sum=('pnl_sum','mean') if 'pnl_sum' in df.columns else ('volume_usd','mean'),
             long_share=('long_share','mean') if 'long_share' in df.columns else ('volume_usd','mean'),
             n_days=('trade_date','nunique'))
        .reset_index())
comp.to_csv("csv_files/agg_by_sentiment.csv", index=False)
print("Saved -> csv_files/agg_by_sentiment.csv")
display(comp)

# plots
if 'pnl_sum' in df.columns and df['pnl_sum'].notna().any():
    ts = df.sort_values('trade_date')
    fig1, ax1 = plt.subplots()
    ax1.plot(ts['trade_date'], ts['pnl_sum'])
    ax1.set_title('Daily PnL Over Time')
    ax1.set_xlabel('Date'); ax1.set_ylabel('PnL (sum)')
    save_plot(fig1, "daily_pnl_timeseries.png"); plt.show()
else:
    print("Skipped PnL timeseries (no PnL column).")

if 'avg_leverage' in df.columns and df['avg_leverage'].notna().any():
    fig2, ax2 = plt.subplots()
    df.boxplot(column='avg_leverage', by='sentiment_clean', ax=ax2)
    ax2.set_title('Average Leverage by Sentiment')
    ax2.set_xlabel('Sentiment'); ax2.set_ylabel('Average Leverage')
    plt.suptitle('')
    save_plot(fig2, "avg_leverage_by_sentiment.png"); plt.show()
else:
    print("Skipped leverage boxplot (no leverage column).")

print(" Core pipeline complete.")

# =======================================================
# === EXTRA ANALYSIS FOR THE REPORT
# =======================================================

# 1) Summary table by sentiment (FIXED metrics list)
metrics = ['n_trades', 'accounts_active', 'volume_usd', 'avg_leverage', 'win_rate', 'pnl_sum', 'long_share']
avail = [m for m in metrics if m in df.columns]
summary = (df.groupby('sentiment_clean')[avail]
             .agg(['mean','median','std','count'])
             .round(4))
summary.to_csv("csv_files/summary_by_sentiment.csv")
display(summary)

# 2) Bar visuals saved to outputs
if 'volume_usd' in df.columns:
    vol = df.groupby('sentiment_clean')['volume_usd'].mean().sort_values()
    fig, ax = plt.subplots(); vol.plot(kind='bar', ax=ax)
    ax.set_title('Average Daily Volume (USD) by Sentiment')
    ax.set_xlabel('Sentiment'); ax.set_ylabel('Avg Volume (USD)')
    save_plot(fig, "avg_volume_by_sentiment.png"); plt.show()

if 'win_rate' in df.columns:
    wr = df.groupby('sentiment_clean')['win_rate'].mean().sort_values()
    fig, ax = plt.subplots(); wr.plot(kind='bar', ax=ax)
    ax.set_title('Average Daily Win Rate by Sentiment')
    ax.set_xlabel('Sentiment'); ax.set_ylabel('Win Rate')
    save_plot(fig, "win_rate_by_sentiment.png"); plt.show()

if 'long_share' in df.columns:
    ls = df.groupby('sentiment_clean')['long_share'].mean().sort_values()
    fig, ax = plt.subplots(); ls.plot(kind='bar', ax=ax)
    ax.set_title('Long Share by Sentiment')
    ax.set_xlabel('Sentiment'); ax.set_ylabel('Long Share')
    save_plot(fig, "long_share_by_sentiment.png"); plt.show()

# 3) Fear vs Greed hypothesis tests
try:
    import scipy.stats as st
    out = []
    def safe_ttest(x, y):
        x = pd.Series(x).dropna(); y = pd.Series(y).dropna()
        if len(x) > 2 and len(y) > 2:
            t,p = st.ttest_ind(x, y, equal_var=False)
            return float(t), float(p), int(len(x)), int(len(y))
        return np.nan, np.nan, len(x), len(y)

    for m in ['volume_usd','win_rate','pnl_sum','avg_leverage','long_share']:
        if m in df.columns:
            t,p,n_fear,n_greed = safe_ttest(
                df.loc[df['sentiment_clean'].str.upper()=='FEAR', m],
                df.loc[df['sentiment_clean'].str.upper()=='GREED', m]
            )
            out.append({'metric':m, 't_stat':t, 'p_value':p, 'n_fear':n_fear, 'n_greed':n_greed})

    tests = pd.DataFrame(out)
    tests.to_csv("csv_files/ttests_fear_vs_greed.csv", index=False)
    display(tests)
except Exception as e:
    print("Skipped t-tests:", e)

# 4) Next-day effect (lag 1)  FIXED duplicate column issue
dfl = df.sort_values('trade_date').copy()
if 'sentiment_code' in dfl.columns:
    dfl['sentiment_code_lag1'] = dfl['sentiment_code'].shift(1)  # idempotent on re-run

    lag_metrics = {}
    for m in ['pnl_sum','win_rate','volume_usd','avg_leverage']:
        if m in dfl.columns:
            lag_metrics[f'nextday_{m}'] = dfl[m]

    lag_df = pd.DataFrame(lag_metrics)
    lag_df['sentiment_code_lag1'] = dfl['sentiment_code_lag1'].values
    lag_df['trade_date'] = dfl['trade_date'].values

    lag_summary = (lag_df.groupby('sentiment_code_lag1', dropna=True)
                   .mean(numeric_only=True)
                   .reset_index())
    lag_summary.to_csv("csv_files/lag1_sentiment_effects.csv", index=False)
    display(lag_summary)
else:
    print("No 'sentiment_code' column; skipping lag analysis.")

# 5) Lightweight linear model (optional; will skip if not enough rows)
try:
    from sklearn.linear_model import LinearRegression
    from sklearn.preprocessing import OneHotEncoder
    from sklearn.compose import ColumnTransformer
    from sklearn.pipeline import Pipeline

    mod_df = df.dropna(subset=['pnl_sum','sentiment_clean']) if 'pnl_sum' in df.columns else None
    if mod_df is not None and len(mod_df) > 10:
        X_num = [c for c in ['n_trades','volume_usd','avg_leverage','long_share','win_rate'] if c in mod_df.columns]
        X_cat = ['sentiment_clean']
        y = mod_df['pnl_sum']

        pre = ColumnTransformer(
            transformers=[
                ('num','passthrough', X_num),
                ('cat', OneHotEncoder(drop='first'), X_cat)
            ])
        pipe = Pipeline([('pre', pre), ('lr', LinearRegression())])
        pipe.fit(mod_df[X_num+X_cat], y)

        cat_levels = pipe.named_steps['pre'].named_transformers_['cat'].get_feature_names_out(X_cat)
        coef_names = X_num + list(cat_levels)
        coefs = pd.DataFrame({'feature':coef_names, 'coef':pipe.named_steps['lr'].coef_})
        coefs.sort_values('coef', ascending=False, inplace=True)
        coefs.to_csv("csv_files/linear_model_pnl_coeffs.csv", index=False)
        display(coefs.head(12))
    else:
        print("Skipped linear model (no pnl_sum or not enough rows).")
except Exception as e:
    print("Skipped linear model:", e)

# 6) Cohorts by account volume (skip if no account field)
if 'volume_usd' in trades.columns and 'account' in trades.columns:
    acct = (trades.groupby(['account','trade_date'])
                  .agg(volume_usd=('volume_usd','sum'),
                       pnl_sum=(col_pnl,'sum') if col_pnl else ('volume_usd','sum'))
                  .reset_index())
    acct = acct.merge(df[['trade_date','sentiment_clean']], on='trade_date', how='left')
    acct['vol_bucket'] = pd.qcut(acct['volume_usd'].rank(method='first'), q=4, labels=['Q1 (low)','Q2','Q3','Q4 (high)'])
    cohort = (acct.groupby(['vol_bucket','sentiment_clean'])
                   .agg(avg_pnl=('pnl_sum','mean'),
                        med_pnl=('pnl_sum','median'),
                        n_days=('trade_date','nunique'))
                   .reset_index())
    cohort.to_csv("csv_files/cohort_volume_by_sentiment.csv", index=False)
    display(cohort)
else:
    print("No account column; skipping cohort analysis.")

# 7) Contrarian day flags
if {'long_share','pnl_sum','sentiment_clean'}.issubset(df.columns):
    contrarian = df[
        ((df['sentiment_clean'].str.upper()=='GREED') & (df['long_share']<0.45)) |
        ((df['sentiment_clean'].str.upper()=='FEAR') & (df['long_share']>0.55))
    ][['trade_date','sentiment_clean','long_share','pnl_sum','volume_usd']]
    contrarian.to_csv("csv_files/contrarian_days.csv", index=False)
    print("Saved -> csv_files/contrarian_days.csv")
    display(contrarian.head(10))

print(" All analyses complete. Artifacts saved in csv_files/ and outputs/.")

# ============================
# Build ds_report.docx from CSVs & charts
# ============================
!pip -q install python-docx

import os, math
import pandas as pd
from docx import Document
from docx.shared import Inches, Pt
from docx.enum.text import WD_ALIGN_PARAGRAPH

CSV_DIR = "csv_files"
OUT_DIR = "outputs"
REPORT_PATH = "ds_report.docx"

def safe_read_csv(path):
    return pd.read_csv(path) if os.path.exists(path) else None

# Load artifacts (some may be None depending on your data)
df_daily      = safe_read_csv(os.path.join(CSV_DIR, "daily_behavior_with_sentiment.csv"))
df_comp       = safe_read_csv(os.path.join(CSV_DIR, "agg_by_sentiment.csv"))
df_summary    = safe_read_csv(os.path.join(CSV_DIR, "summary_by_sentiment.csv"))
df_tests      = safe_read_csv(os.path.join(CSV_DIR, "ttests_fear_vs_greed.csv"))
df_lag        = safe_read_csv(os.path.join(CSV_DIR, "lag1_sentiment_effects.csv"))
df_cohort     = safe_read_csv(os.path.join(CSV_DIR, "cohort_volume_by_sentiment.csv"))
df_contrarian = safe_read_csv(os.path.join(CSV_DIR, "contrarian_days.csv"))

IMG_PNL    = os.path.join(OUT_DIR, "daily_pnl_timeseries.png")
IMG_VOL    = os.path.join(OUT_DIR, "avg_volume_by_sentiment.png")
IMG_WR     = os.path.join(OUT_DIR, "win_rate_by_sentiment.png")
IMG_LONG   = os.path.join(OUT_DIR, "long_share_by_sentiment.png")
IMG_LEV    = os.path.join(OUT_DIR, "avg_leverage_by_sentiment.png")  # may not exist if no leverage

doc = Document()

def title(text):
    p = doc.add_heading(text, 0)
    p.alignment = WD_ALIGN_PARAGRAPH.CENTER

def h1(text): doc.add_heading(text, level=1)
def h2(text): doc.add_heading(text, level=2)

def add_paragraph(text, bold=False):
    p = doc.add_paragraph()
    run = p.add_run(text)
    if bold: run.bold = True
    p.paragraph_format.space_after = Pt(6)
    return p

def add_bullets(items):
    for it in items:
        doc.add_paragraph(str(it), style='List Bullet')

def add_table_from_df(df: pd.DataFrame, caption: str, max_rows=15):
    if df is None or df.empty:
        add_paragraph(f"(No data found for {caption}.)")
        return
    add_paragraph(caption, bold=True)
    df_show = df.copy()
    if len(df_show) > max_rows:
        df_show = df_show.head(max_rows)
        add_paragraph(f"Showing first {max_rows} rows.", bold=False)
    table = doc.add_table(rows=1, cols=len(df_show.columns))
    hdr = table.rows[0].cells
    for j, c in enumerate(df_show.columns):
        hdr[j].text = str(c)
    for _, row in df_show.iterrows():
        cells = table.add_row().cells
        for j, c in enumerate(df_show.columns):
            val = row[c]
            if isinstance(val, float) and (abs(val) > 1e6 or abs(val) < 1e-3) and val != 0:
                cells[j].text = f"{val:.4e}"
            elif isinstance(val, float):
                cells[j].text = f"{val:.4f}"
            else:
                cells[j].text = str(val)
    doc.add_paragraph("")  # spacing

def add_image(path, caption, width_in=6.2):
    if os.path.exists(path):
        doc.add_paragraph(caption, style=None)
        doc.add_picture(path, width=Inches(width_in))
        doc.add_paragraph("")  # spacing

# ===== Title + Executive Summary =====
title("Analysis of Trader Behavior vs Market Sentiment  Mohammed Shahrukh")

add_paragraph("Objective", bold=True)
add_paragraph(
    "Explore how trader performance (P&L, win rate, positioning, volume) aligns or diverges "
    "from overall market sentiment (FearGreed), and surface signals that can guide trading decisions."
)

# Compute a few headline stats (guarded for missing data)
bullets = []
if df_comp is not None and not df_comp.empty:
    # Try to compute deltas (Fear vs Greed) where possible
    comp = df_comp.copy()
    # Normalize sentiment column name
    sent_col = [c for c in comp.columns if c.lower().startswith('sentiment')][0]
    comp[sent_col] = comp[sent_col].astype(str)
    def get_metric(sent_label, metric):
        row = comp.loc[comp[sent_col].str.upper()==sent_label.upper()]
        if not row.empty and metric in comp.columns:
            return float(row.iloc[0][metric])
        return None

    fear_vol  = get_metric('FEAR', 'volume_usd')
    greed_vol = get_metric('GREED','volume_usd')
    if fear_vol is not None and greed_vol is not None:
        vol_diff = (greed_vol - fear_vol) / max(1e-9, fear_vol)
        bullets.append(f"Avg daily volume on GREED vs FEAR: {vol_diff:+.1%} difference.")

    fear_wr  = get_metric('FEAR', 'win_rate')
    greed_wr = get_metric('GREED','win_rate')
    if fear_wr is not None and greed_wr is not None:
        wr_diff = (greed_wr - fear_wr)
        bullets.append(f"Win-rate shift (GREED - FEAR): {wr_diff:+.2%}.")

    # Which sentiment had the highest average PnL?
    if 'pnl_sum' in comp.columns and comp['pnl_sum'].notna().any():
        top = comp.loc[comp['pnl_sum'].idxmax()]
        bullets.append(f"Highest avg daily PnL sentiment: {top[sent_col]} "
                       f"({top['pnl_sum']:.2f}).")

if df_tests is not None and not df_tests.empty:
    sig = df_tests[(df_tests['p_value'] < 0.05)]
    if not sig.empty:
        bullets.append("Statistically significant FEAR vs GREED differences detected for: " +
                       ", ".join(sig['metric'].tolist()) + " (p<0.05).")

if not bullets:
    bullets.append("No strong differences surfaced yet (limited sample or missing fields).")

h1("Executive Summary")
add_bullets(bullets)

# ===== Data & Method =====
h1("Data & Method")
add_paragraph("Datasets", bold=True)
add_bullets([
    "Hyperliquid historical trades (daily aggregated).",
    "Bitcoin Fear & Greed Index (multi-class sentiment: Neutral, Fear, Greed, Extreme Greed)."
])
add_paragraph("Methodology", bold=True)
add_bullets([
    "Normalized and parsed timestamps; bucketed trades to UTC days.",
    "Engineered features: volume_usd, win_rate, long_share, pnl_sum (where available).",
    "Merged daily aggregates with daily sentiment.",
    "Evaluated cross-sectional differences and simple next-day (lag-1) relationships.",
])

# ===== Findings =====
h1("Findings")

h2("Summary by Sentiment")
add_table_from_df(df_comp, "Average metrics by sentiment (from agg_by_sentiment.csv)")
if df_summary is not None:
    add_table_from_df(df_summary, "Summary stats (mean/median/std) by sentiment")
else:
    add_paragraph("(summary_by_sentiment.csv not found)")

h2("Key Visuals")
add_image(IMG_PNL,  "Figure: Daily PnL Over Time")
add_image(IMG_VOL,  "Figure: Average Daily Volume (USD) by Sentiment")
add_image(IMG_WR,   "Figure: Average Daily Win Rate by Sentiment")
add_image(IMG_LONG, "Figure: Long Share by Sentiment")
add_image(IMG_LEV,  "Figure: Average Leverage by Sentiment")

h2("Hypothesis Tests: FEAR vs GREED")
add_table_from_df(df_tests, "Two-sample t-tests on FEAR vs GREED (Welch)")

h2("Next-Day (Lag-1) Effects")
add_table_from_df(df_lag, "Mean of next-day metrics grouped by prior-day sentiment code")

if df_cohort is not None:
    h2("Cohorts: Volume Buckets  Sentiment")
    add_table_from_df(df_cohort, "Average/Median PnL by volume quartile and sentiment")
if df_contrarian is not None:
    h2("Contrarian Days")
    add_table_from_df(df_contrarian.head(20), "Sample of contrarian positioning days (first 20 rows)")

# ===== Recommendations =====
h1("Recommendations")
recs = [
    "Tune risk by sentiment: reduce exposure on lower-edge days (based on PnL/WR differences), press size when evidence favors GREED or specific cohorts.",
    "Watch transitions: NeutralGreed or GreedExtreme Greed; confirm with volume and direction bias (long_share).",
    "Run a small, rules-based backtest (position sizing + stop logic) using these signals before production.",
]
add_bullets(recs)

# ===== Limitations & Next Steps =====
h1("Limitations & Next Steps")
add_bullets([
    "Leverage field may be missing; proxies or exchange-level leverage data could refine risk insights.",
    "PnL depends on how trades were closed; adding unrealized PnL could improve coverage.",
    "More robust modeling (regularization, time-aware splits) and proper backtests recommended.",
])

doc.save(REPORT_PATH)
print(f" Report created: {REPORT_PATH}")


import shutil

# Zip csv_files folder
shutil.make_archive('csv_files', 'zip', 'csv_files')

# Zip outputs folder
shutil.make_archive('outputs', 'zip', 'outputs')


from google.colab import files
files.download('csv_files.zip')
files.download('outputs.zip')






```
